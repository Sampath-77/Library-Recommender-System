```python
books_data = books_data[['title','Authors','description','Category','Publisher']]
books_data
```

```python
books_data['Description'] = books_data['Authors']+books_data['description']+books_data['Category']+books_data['Publisher']
books_data
```

```python
books_data = books_data.drop(columns=['Authors','description','Category','Publisher'])
books_data
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

books_data.head()

def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text, re.I)
    text = text.lower()
      
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    return " ".join(tokens)

books_data['processed_description'] = books_data['Description'].apply(preprocess_text)
books_data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np


tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(books_data['processed_description'])

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend_books(book_title, cosine_sim=cosine_sim):
    idx = books_data[books_data['title'] == book_title].index[0]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    
    book_indices = [i[0] for i in sim_scores]
    return books_data['title'].iloc[book_indices]


recommended_books = recommend_books("Journey Through Heartsongs")
print("Recommended books:\n", recommended_books)
```
